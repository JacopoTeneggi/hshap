```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
import numpy as np
import pandas as pd
from tqdm import tqdm

exp_mapper = ["hexp/absolute_0", "gradexp", "deepexp", "partexp", "lime"]
LOR_df_clips = []
for exp_name in tqdm(exp_mapper):
    exp_df = pd.read_csv(os.path.join("LOR", exp_name, "results.csv"))
    LOR_df_clips.append(exp_df)
LOR_df = pd.concat(LOR_df_clips, axis=0)
LOR_df["perturbation_size"] /= 1200*1600
```

```python
sns.set_theme()
sns.set_context("paper")

g = sns.lineplot(data=LOR_df, x="perturbation_size", y="logit", hue="exp_name", legend=False)
g.set(xlabel="Fraction of image ablated", ylabel="Logit output as a function of original logit")
plt.savefig("LOR_seaborn.eps")
plt.savefig("LOR_seaborn.jpg")
plt.savefig("LOR_seaborn.pdf")
```
