```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
import numpy as np
import pandas as pd
from tqdm import tqdm

c = [6]
exp_mapper = ["hexp/absolute_0", "gradexp", "deepexp", "partexp", "lime"]
LOR_df_clips = []
for j, n in enumerate(c):
    for exp_name in tqdm(exp_mapper):
        exp_df = pd.read_csv(os.path.join("LOR", exp_name, f"results_{n}.csv"))
        LOR_df_clips.append(exp_df)
LOR_df = pd.concat(LOR_df_clips, axis=0)
LOR_df["perturbation_size"] /= 100*120
```
